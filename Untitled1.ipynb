```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
url = 'https://raw.githubusercontent.com/cs109/2014_data/master/diamonds.csv';
diamonds=pd.read_csv(url,sep=',',index_col=0)
diamonds.head()
diamonds.dimen
```

```python
diamonds.describe()
```

```python
diamonds['price'].mean()
diamonds['price'].hist(bins=50,color='black')
plt.title('Distribution of Price')
plt.xlabel('Price')
```

```python
#Average distribution price 
mean=diamonds['price'].mean()
mean
```

```python
#Changing bins size from 20 to 500. What do you notice? 
diamonds['carat'].hist(bins=10,color='red')
plt.title(' Distribution of weights in carats')
plt.xlim(0,3)
plt.xlabel('weights in carats ')
diamonds.shape
```

Plot the density of the price of the diamonds

```python
ax=diamonds['price'].plot.kde()
plt.title('Distribution of Price')
xlabel("Price")
```

```python
#Finding the correlation between two variable price and carat 
diamonds['price'].corr(diamonds['carat'])
```

```python
diamonds.plot(x='carat', y='price', kind = 'scatter', color = 'black', alpha = 1)
```

We can also create a scatter plot using matplotlib.pyplot instead of pandas directly.

```python
plt.scatter(diamonds['carat'], diamonds['price'], color = 'black', alpha = 0.05)
plt.xlabel('Carat')
plt.ylabel('Price')
```

```python
diamonds.groupby('color').plot(x='carat', y='price', kind = 'scatter', color = 'black', alpha = 1)
```
